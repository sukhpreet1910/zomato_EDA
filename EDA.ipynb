## Right resion to drop these tables

```python
columns_to_drop = ['url', 'phone', 'address', 'reviews_list', 'menu_item']
```

```python
df = df.drop(columns=columns_to_drop)
```

```python
df.head()
```

# Data Cleaning

### Dropping Duplicates

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

### Handling values in Rate Column

```python
df['rate'].unique()
```

#### As we see, it has some values 'NEW', '-' and we need to get rid of /5 for further analysis on this series

```python
def handle_rate(value):
    if value == 'NEW' or value == '-':
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handle_rate)
df['rate'].head()
```

```python
df['rate'].isnull().sum()
```

#### Fill null values with mean

```python
# Set the option to opt-in to the future behavior
pd.set_option('future.no_silent_downcasting', True)

# Your existing code
df['rate'] = df['rate'].fillna(df['rate'].mean())
```

```python
df['rate'].unique()
```

```python
df.info()
```

## Handle Null Values

#### Dish_Liked columns have too many null values, it's better to drop while column

```python
df = df.drop('dish_liked', axis = 1)
```

```python
df.info()
```

#### Dropping Null Values from every other column

```python
df.dropna(inplace=True)
```

```python
df.info()
```

#### Our data set does not have duplicates or null values anymore. Now we can start analysing each column

```python
df.rename(columns = {'approx_cost(for two people)': 'cost_2people', 'listed_in(type)': 'type'}, inplace=True)
```

```python
df.head()
```

#### Let's see the difference between location and listed_in(city)

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

#### Both the columns are presenting same kind of values, Let's keep only one

```python
df = df.drop('listed_in(city)', axis = 1)
```

```python
df.head()
```

```python
df['cost_2people'].unique()
```

#### As we can see, Some values in cost column inclused ',' in them, That means its not of integer type. But cost always should of int or float type

```python
df['cost_2people'].dtype
```

```python
def handle_cost(value):
    if ',' in value:
        value = str(value).replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['cost_2people'] = df['cost_2people'].apply(handle_cost)
```

```python
df['cost_2people'].unique()
```

```python
df.head(100)
```

#### Let's see whats in rest_type column

```python
df['rest_type'].value_counts()
```

#### It shows, there are some types which have very less number of rastaurant.  There are 19010 rastaurants which is of type Quick Bites. On the other hand there are some types which have few hundered rastaurants. Let's name it as "Others"

```python
rest_type = df['rest_type'].value_counts()
rest_type_below300 = rest_type[rest_type < 1000]
```

```python
def handle_rest_type(value):
    if value in rest_type_below300:
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

### Let's Look at Location column

```python
df['location'].value_counts()
```

#### Making a cluster of locations having less than 300 rastaurants

```python
locations = df['location'].value_counts()
locations_less_than_300 = locations[locations < 300]
```

```python
def handle_locations(value):
    if value in locations_less_than_300:
        return str('others')
    else:
        return value
    
df['location'] = df['location'].apply(handle_locations)
df['location'].value_counts()
```

```python
df.head()
```

### Let's Look at Cuisines column

```python
df['cuisines'].value_counts()
```

#### Making a clusters of cuisines having less than 100 rastaurants

```python
cuisines = df['cuisines'].value_counts()
cuisines_less_than_100 = cuisines[cuisines < 100]
```

```python
def handle_cuisines(value):
    if value in cuisines_less_than_100:
        return 'others'
    else:
        return value
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

#### Let's look at type column

```python
df['type'].value_counts()
```

#### This column has only few unique values and all the unique values have good amount of rastaurants. So there is no need to do any changes in this data

```python
df.head()
```

# Data is clean, Now we can start visualising it

## Count plot of various locations

```python
sns.set_theme(style="darkgrid")
custom_palette = sns.color_palette("Set2")
plt.figure(figsize=(16, 6))
sns.countplot(x="location", data=df, order=df['location'].value_counts().index, palette='deep', legend= False, hue="location")
plt.title('Countplot of Restaurants by Location')
plt.xticks(rotation = 90)
plt.show()
```

### Visualising online order

```python
plt.figure(figsize = (8, 6))
sns.countplot(x = 'online_order', data = df, palette='deep', legend= False, hue="online_order")
plt.title("Counplot of rastaurants having online order facility")
plt.xticks(rotation = 90)
plt.show()
```

### Visualising Book Table

```python
plt.figure(figsize= (6, 6))
sns.countplot(x = 'book_table', data = df, order = df['book_table'].value_counts().index, palette='deep', legend= False, hue="book_table")
plt.title("Counplot of rastaurants having Book Table facility")
plt.xticks(rotation = 90)
plt.show()
```

## Visualising online order vs rating
